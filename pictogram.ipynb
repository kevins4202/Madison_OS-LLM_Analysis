```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.patches as mpatches
import pywaffle
import imageio
import cv2
%matplotlib inline
```

```python
exam= pd.read_csv(os.getcwd()+f'/exams/18-spring-mid.csv')
```

```python
exam
```

```python
plt.figure(figsize = (300,18))
for i in range(150):
    for j in range(2,6):
        if(exam.iat[i,j] == 1):
            img = cv2.imread("check.png")
            
        else: 
            img = cv2.imread("white.png")
            
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB), extent=[i+0.5, i+1.5, (j-2), (j-1)])
        plt.xlim(0, 32)
        plt.ylim(0, 3)
#         plt.show()
plt.xticks(np.arange(1,152,1), fontsize = 36)
ax = plt.gca()
xticks = ax.xaxis.get_major_ticks()
xticks[-1].set_visible(False)
plt.yticks(range(0,4),["GPT3.5", "Bard", "GPT-4", "Gemini"], fontsize = 36)
plt.savefig('18-spring-mid3.png')
```

```python
exams = ['18-spring-final.csv','18-spring-mid.csv','21-fall-final.csv', '21-fall-mid.csv']

for ex in exams:
    exam= pd.read_csv(os.getcwd()+f'/exams/{ex}')
    
    plt.figure(figsize = (exam.shape[0]+2,18))
    for i in range(exam.shape[0]):
        for j in range(2,6):
            if(exam.iat[i,j] == 1):
                img = cv2.imread("check.png")

            else: 
                img = cv2.imread("white.png")

            plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB), extent=[i+0.5, i+1.5, (j-2), (j-1)])
            plt.xlim(0, 32)
            plt.ylim(0, 3)
    #         plt.show()
    plt.xticks(np.arange(1,exam.shape[0]+2,1), fontsize = 36)
    ax = plt.gca()
    xticks = ax.xaxis.get_major_ticks()
    xticks[-1].set_visible(False)
    plt.yticks(range(0,4),["GPT3.5", "Bard", "GPT-4", "Gemini"], fontsize = 36)
    plt.savefig(f'{ex.split(".")[0]}-comparison.png')

```

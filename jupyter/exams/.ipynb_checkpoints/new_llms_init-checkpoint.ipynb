```python
for exam in new_exams:
    df = pd.read_csv(exam)
    df = df.drop("Unnamed: 0", axis = 1) if "Unnamed: 0" in df.columns else df

    for col in df.columns[:-1]:
        df[col] = (df[col] == df["Answer"]).astype(int)

    df = df.drop("Answer", axis = 1)
    df = df.rename({col:col.lower() + "_pts_earned" for col in df.columns}, axis = 1)

    df.to_csv(exam, index = False)
```

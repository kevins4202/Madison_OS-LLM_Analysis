```python
import pandas as pd
```

```python
new_exams = ["18-spring-final.csv", "18-spring-mid.csv","21-fall-final.csv","21-fall-mid.csv"]
```

```python
for exam in new_exams:
    df1 = pd.read_csv(exam)
    df2 = pd.read_csv(exam[:-4]+"-new.csv")

    df3 = pd.concat([df1, df2], axis = 1)
    # print(df3)

    df3.to_csv(exam[:-4]+"_combined.csv", index = False)
```
